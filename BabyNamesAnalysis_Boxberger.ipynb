#Identify any name that was in the top 3 for a year (male) and then clean it up

```python
top_3_male_df=national_names_male_df.loc[national_names_male_df["Rank"]<=3]
top_3_male = top_3_male_df["Name"].unique()
top_3_male_df = pd.DataFrame(top_3_male)
top_3_male_df
```

#Plot male names that were top 3 to see how they look

```python
sns.set(rc={'figure.figsize':(11,8.5)})

sns.axes_style()

plt.title("Baby Name Frequency over Time",fontsize = 20)
plt.xlabel("Year",fontsize = 20)
plt.ylabel("Count of Name",fontsize = 20)

labels=[]
for name in top_3_male:
    plt.plot(national_names_male_df.loc[national_names_male_df["Name"]==name]["Year"],
             national_names_male_df.loc[national_names_male_df["Name"]==name]["Count"])
    labels.append(name)

plt.legend(labels, loc='upper right', 
           bbox_to_anchor=[1.3,1], fontsize=14)

plt.show()
```

#Create plotting functions (male and female) for ease in analysis

```python
def NamePlotMale (names,title):
    sns.set(rc={'figure.figsize':(11,8.5)})

    sns.axes_style()

    plt.title("Male Baby Name Frequency over Time",fontsize = 20)
    plt.xlabel("Year",fontsize = 20)
    plt.ylabel("Count of Name",fontsize = 20)

    labels=[]
    for name in names:
        plt.plot(national_names_male_df.loc[national_names_male_df["Name"]==name]["Year"],
                 national_names_male_df.loc[national_names_male_df["Name"]==name]["Count"])
        labels.append(name)

    plt.legend(labels, loc='upper right', bbox_to_anchor=[1.15,1], fontsize=14)
    plt.savefig (title +'.png')
    plt.show()
```

```python
def NamePlotFemale (names,title):
    sns.set(rc={'figure.figsize':(11,8.5)})

    sns.axes_style()

    plt.title("Female Baby Name Frequency over Time",fontsize = 20)
    plt.xlabel("Year",fontsize = 20)
    plt.ylabel("Count of Name",fontsize = 20)

    labels=[]
    for name in names:
        plt.plot(national_names_female_df.loc[national_names_female_df["Name"]==name]["Year"],
                 national_names_female_df.loc[national_names_female_df["Name"]==name]["Count"])
        labels.append(name)

    plt.legend(labels, loc='upper right', bbox_to_anchor=[1.15,1], fontsize=14)
    plt.savefig (title +'.png')
    plt.show()
```

#I did a lot of data manipulation (above) to get state names and how they compared to national names,
and then created a plotting function to compare state to national.  But ultimately I wasn't gleaning what
I wanted from the data.  A teammate was taking a different approach with the state data, so I abandoned my
pursuits. 

```python
def NamePlotStateMale (names,states,title):
    sns.set(rc={'figure.figsize':(11,8.5)})

    sns.axes_style()

    plt.title("Male Baby Name Frequency over Time",fontsize = 20)
    plt.xlabel("Year",fontsize = 20)
    plt.ylabel("Count of Name",fontsize = 20)

#National
    labels=[]
    for name in names:
        
#        plt.plot(national_names_male_df.loc[national_names_male_df["Name"]==name]["Year"],
#                 national_names_male_df.loc[national_names_male_df["Name"]==name]["Count"])
        labels.append(name)

#State
    for name in names:
        for state in states:
            
            plt.plot(state_names_male_df.loc[(state_names_male_df["Name"]==name) &
                    (state_names_male_df["State"]==state)]["Year"],
                     state_names_male_df.loc[(state_names_male_df["Name"]==name) &
                    (state_names_male_df["State"]==state)]["Count"])
        
    plt.legend(labels, loc='upper right', bbox_to_anchor=[1.15,1], fontsize=14)
    plt.savefig (title +'.png')
    plt.show()
```
